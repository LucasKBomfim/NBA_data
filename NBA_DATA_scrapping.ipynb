```python
import pandas as pd
import requests 
pd.set_option('display.max_columns',None) # Show all the columns in a dataset
import time
import numpy as np
```

```python
url_inicial = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season=2021-22&SeasonType=Regular%20Season&StatCategory=PTS'
r = requests.get(url_inicial).json()
table_headers = r['resultSet']['headers']

df_cols = ['Year' , 'Season_type'] + table_headers
df = pd.DataFrame(columns = df_cols)

df

headers = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
}

season_types = ['Regular%20Season','Playoffs']
years = ['1946-47','1947-48','1948-49','1949-50','1950-51','1951-52','1952-53','1953-54','1954-55','1955-56','1956-57','1957-58','1958-59','1959-60','1960-61','1961-62','1962-63','1963-64','1964-65','1965-66','1966-67','1968-69','1969-70','1970-71','1971-72','1972-73','1973-74',
'1974-75','1975-76','1976-77','1977-78','1978-79','1979-80','1980-81','1981-82','1982-83','1983-84','1984-85','1985-86','1986-87','1987-88','1988-89','1989-90','1990-91','1991-92','1992-93','1993-94','1994-95','1995-96','1996-97','1997-98','1998-99','1999-00','2000-01',
'2000-01','2001-02','2002-03','2003-04','2004-05','2005-06','2006-07','2007-08','2008-09','2009-10','2010-11','2011-12','2012-13','2013-14','2014-15','2015-16','2016-17','2017-18','2018-19','2019-20','2020-21','2021-22']

begin_loop = time.time()

for y in years:
    for s in season_types:
        api_url = 'https://stats.nba.com/stats/leagueLeaders?LeagueID=00&PerMode=Totals&Scope=S&Season='+y+'&SeasonType='+s+'&StatCategory=PTS'
        print(api_url)
        r = requests.get(url=api_url,headers = headers).json()
        temp_df1 = pd.DataFrame(r['resultSet']['rowSet'],columns = table_headers)
        temp_df2 = pd.DataFrame({'Year' : [y for i in range(len(temp_df1))], 
                                'Season_type' : [s for i in range(len(temp_df1))]})
        temp_df3 = pd.concat([temp_df2,temp_df1], axis = 1)
        df = pd.concat([df,temp_df3], axis = 0)
        print(f'Finished scrapping data for the {y} {s}')
        lag = np.random.uniform(30,100)
        print(f'...waiting {round(lag,1)} seconds')
        time.sleep(lag)
        
print(f'Process completed! Total run time: {round((time.time()-begin_loop)/60,2)} minutes')
```

```python
df.to_csv('NBA_Data.csv')
```
